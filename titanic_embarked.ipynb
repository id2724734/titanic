```python
import pandas as pd
%matplotlib inline
import seaborn as sns
```

# Load Dataset

```python
train= pd.read_csv("train.csv", index_col = "PassengerId")
print(train.shape)
train.head()
```

```python
test = pd.read_csv("test.csv", index_col="PassengerId")
#세로가로 보여주기
print(test.shape)
test.head()
```

# Preprocessing(전처리)

```python
#text를 숫자로 바꿔주는 것
#Encode Sex 

#Fill in missing Fare -> 빈 값 채워주기
```

## Encode Sex

```python
train["Sex_encode"] = train["Sex"].replace("male",0).replace("female",1)

print(train.shape)
train[["Sex", "Sex_encode"]].head()
```

```python
#train 데이터 했으면 test 데이터도 똑같이
test["Sex_encode"] = test["Sex"].replace("male",0).replace("female",1)
print(test.shape)
test[["Sex","Sex_encode"]].head()
```

# Encode Embarked

```python
# C == 0, S == 1, Q == 2
# 이렇게 되면 파이썬에서는
# 2 * S == Q?
# S + S = Q??
# 이렇게 읽을 수 있는 가능성이 있음 그래서 이럴때는

# One Hot Encoding
# [1,0,0] [0.1,0] [0,0,1]
#파이썬에서는 True == 1, False == 0
# C == [1,0,0] ->  [True, False, False]
# S == [0,1,0] -> [False, True, False]
# Q == [0,0,1] -> [False, False, True]

train["Embarked_C"] = train["Embarked"] == "C"
train["Embarked_S"] = train["Embarked"] == "S"
train["Embarked_Q"] = train["Embarked"] == "Q"

print(train.shape)
train[["Embarked","Embarked_C","Embarked_S","Embarked_Q"]].head()
```

```python
test["Embarked_C"] = test["Embarked"] == "C"
test["Embarked_S"] = test["Embarked"] == "S"
test["Embarked_Q"] = test["Embarked"] == "Q"

print(test.shape)
test[["Embarked","Embarked_C","Embarked_S","Embarked_Q"]].head()
```

```python
# SibSp + Parch = FamilySize

train["FamilySize"] = train["SibSp"] + train["Parch"] + 1
print(train.shape)

train[["SibSp","Parch","FamilySize"]].head(10)
```

```python
# 싱글 -> 나 혼자
# 핵가족 -> 2명이상 5미만
# 대가족 -> 5명 이상
train.loc[train["FamilySize"] == 1, "FamilyType"] = "SIngle"
train.loc[(train["FamilySize"] > 1) & (train["FamilySize"] < 5), "FamilyType"] = "Nuclear"
train.loc[train["FamilySize"] > 4, "FamilyType"] = "Big"

print(train.shape)

train[["FamilySize", "FamilyType"]].head(10)

train["F_Size_S"] = train["FamilyType"] == "Single"
train["F_Size_N"] = train["FamilyType"] == "Nuclear"
train["F+Size_B"] = train["FamilyType"] == "Big"
```

```python
train["F_Size_S"] = train["FamilyType"] == "Single"
train["F_Size_N"] = train["FamilyType"] == "Nuclear"
train["F_Size_B"] = train["FamilyType"] == "Big"

print(train.shape)

train[["FamilySize","F_Size_S","F_Size_N","F_Size_B"]].head(10)
```

```python
# SibSp + Parch = FamilySize

test["FamilySize"] = test["SibSp"] + test["Parch"] + 1

print(test.shape)

test[["SibSp","Parch","FamilySize"]].head(10)
```

```python
# 싱글 -> 나 혼자
# 핵가족 -> 2명이상 5미만
test.loc[test["FamilySize"] == 1, "FamilyType"] = "SIngle"
test.loc[(test["FamilySize"] > 1) & (test["FamilySize"] < 5), "FamilyType"] = "Nuclear"
test.loc[test["FamilySize"] > 4, "FamilyType"] = "Big"

print(test.shape)

test[["FamilySize", "FamilyType"]].head(10)
```

```python
test["F_Size_S"] = test["FamilyType"] == "Single"
test["F_Size_N"] = test["FamilyType"] == "Nuclear"
test["F_Size_B"] = test["FamilyType"] == "Big"

print(test.shape)

test[["FamilySize","F_Size_S","F_Size_N","F_Size_B"]].head(10)
```

```python
train["Pclass_F"] = train["Pclass"] == 1
train["Pclass_B"] = train["Pclass"] == 2
train["Pclass_E"] = train["Pclass"] == 3

print(train.shape)
train[["Pclass","Pclass_F","Pclass_B","Pclass_E"]]
```

```python
test["Pclass_F"] = test["Pclass"] == 1
test["Pclass_B"] = test["Pclass"] == 2
test["Pclass_E"] = test["Pclass"] == 3

print(test.shape)
test[["Pclass","Pclass_F","Pclass_B","Pclass_E"]]
```

# Fill in missing Fare

```python
# class 당 평균 나이값 채우기
mean_age_by_pclass = train.groupby("Pclass")["Age"].mean()
mean_age_by_pclass
```

```python
train["Age(fill)"] =train["Age"]
print(train.shape)

train[["Pclass","Age","Age(fill)"]].head(30)
```

```python
train.loc[(train["Pclass"] == 1) & (train["Age"].isnull()), "Age(fill)"] = mean_age_by_pclass.loc[1]
train.loc[(train["Pclass"] == 2) & (train["Age"].isnull()), "Age(fill)"] = mean_age_by_pclass.loc[2]
train.loc[(train["Pclass"] == 3) & (train["Age"].isnull()), "Age(fill)"] = mean_age_by_pclass.loc[3]

print(train.shape)

train[["Pclass", "Age", "Age(fill)"]].head(30)
```

```python
train.loc[train["Age(fill)"] < 15, "AgeType"] = "Young"
train.loc[(train["Age(fill)"] >= 15) & (train["Age"] < 30), "AgeType"] = "Medium"
train.loc[train["Age(fill)"] >= 30, "AgeType"] = "Old"

train["Age_Y"] = train["AgeType"] == "Young"
train["Age_M"] = train["AgeType"] == "Medium"
train["Age_O"] = train["AgeType"] == "Old"

print(train.shape)

train[["AgeType","Age_Y","Age_M","Age_O"]].head(30)
```

```python
test.loc[test["Age(fill)"] < 15, "AgeType"] = "Young"
test.loc[(test["Age(fill)"] >= 15) & (test["Age(fill)"] < 30), "AgeType"] = "Medium"
test.loc[test["Age(fill)"] >= 30, "AgeType"] = "Old"

test["Age_Y"] = test["AgeType"] == "Young"
test["Age_M"] = test["AgeType"] == "Medium"
test["Age_O"] = test["AgeType"] == "Old"

print(test.shape)

test[["Age","Age(fill)","AgeType","Age_Y","Age_M","Age_O"]].head(30)
```

```python
#train 값은 요금 값에 빈 값이 없음
#test["Fare"]로 한번 묶어주는 이유는
#전체가 다 새로운 데이터로 업뎃으로 수정을 해야 함
#그냥 원본 데이터를 막 바꾸는걸 꺼려하게 만들어 져 있기 때문에
#그리고 값을 0으로 한 이유는 빈 값이 1개 값이라서 힘들일 필요가 없다고 생각이 들어서


test[test["Fare"].isnull()]
```

# Train


```python
# Feature
# 답을 맞추는 데 필요한 col ( Pclass, Sex, Fare )
# Lable
# 맞춰야 될 값 (Survived)

# Feature 와 Lable 이 있음 트리를 만들 수 있다.
# Decision Tree
```

```python
feature_names = ["Pclass_F","Pclass_B","Pclass_E", "Sex_encode", "Fare","Embarked_C","Embarked_S","Embarked_Q","F_Size_S","F_Size_N","F_Size_B","Age_Y","Age_M","Age_O"]
feature_names
```

```python
X_train = train[feature_names]
print(X_train.shape)
X_train.head()
```

```python
X_test = test[feature_names]

print(X_test.shape)
X_test.head()
```

#### Decision Tree에는 text가 들어가지 않는다

```python
label_name = "Survived"

y_train = train[label_name]

print(y_train.shape)
y_train.head()

#test는 survived 없음 우리가 맞춰야 되서 안해도 됨 
#수학적으로 vector는 소문자 matrix 대문자
#요즘은 상관 없음
```

# Use Decision Tree

```python
# scikit-learn == sklearn
from sklearn.tree  import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth = 10)
model
```

```python
# 1)fit 머신러닝 알고리즘을 학습 시키는 것 데이터 사이의 패턴을 머신러닝 알고리즘이 스스로 판단
# fit 할때는 train 데이터 predict 할때는 test 데이터
# 2) predict
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)

print(prediction.shape)
prediction[0:5]
```

# Submit

```python
submit= pd.read_csv("gender_submission.csv", index_col = "PassengerId")

submit["Survived"] = prediction

print(submit.shape)
submit.head()
```
